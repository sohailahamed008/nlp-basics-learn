```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
nltk.download('punkt_tab')
nltk.download('wordnet')
```

```python
df=pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['review_lowercase']=df['Review'].str.lower()
```

```python
df.head()
```

```python
en_stopwords= stopwords.words("english")
```

```python
en_stopwords.remove('not')
```

```python
df['review_no_stopwords']=df['review_lowercase'].apply(lambda x:' '.join(word for word in x.split() if word not in (en_stopwords) ))
```

```python
df['review_no_stopwords'][0]
```

```python
df['review_no_stp_no_pun']=df['review_no_stopwords'].apply(lambda x: re.sub(r"[*]","star",x))
```

```python
df.head()
```

```python
df['review_no_stp_no_pun']=df['review_no_stp_no_pun'].apply(lambda x: re.sub(r"([^\w\s])","",x))
```

```python
df.head()
```

```python
df['review_no_stp_no_pun'][0]
```

```python
df['tokenize']=df['review_no_stp_no_pun'].apply(word_tokenize)
```

```python
df['tokenize'][0]
```

```python
ps=PorterStemmer()
```

```python
df['stemmed']=df['tokenize'].apply(lambda tokens:[ps.stem(token) for token in tokens])
```

```python
df.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
df['lemmatized']=df['tokenize'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
df['lemmatized'][0]
```

```python
tokens_clean=sum(df['lemmatized'],[])
```

```python
unigrams=(pd.Series(nltk.ngrams(tokens_clean,1)).value_counts())
print(unigrams)
```

```python
bigrams=(pd.Series(nltk.ngrams(tokens_clean,2)).value_counts())
print(bigrams)
```

```python
trigrams=(pd.Series(nltk.ngrams(tokens_clean,3)).value_counts())
print(trigrams)
```
